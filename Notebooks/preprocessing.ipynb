# Preprocessing

### Filter Data

```python
# Drop data from game_results for tournaments before 2008
game_results = game_results.copy()[game_results['YEAR'] >= 2008]
game_results
```

### Make Naming Conventions Consistent

```python
# Strip any leading or trailing whitespace from Team Name features
team_stats['TEAM'] = team_stats['TEAM'].apply(lambda name: name.strip())
game_results['WTEAM'] = game_results['WTEAM'].apply(lambda name: name.strip())
game_results['LTEAM'] = game_results['LTEAM'].apply(lambda name: name.strip())
```

```python
# Replace abbreviation "St." with either "Saint" or "State" where appropriate
team_stats['TEAM'] = team_stats['TEAM'].apply(lambda name: name.replace("St.", "Saint") if name in ["St. Bonaventure", "Mount St. Mary's", "St. John's"] else name)
team_stats['TEAM'] = team_stats['TEAM'].apply(lambda name: name.replace("St.", "State") if "St." in name else name)

game_results['WTEAM'] = game_results['WTEAM'].apply(lambda name: name.replace("St.", "State") if name == "Murray St." else name)
game_results['WTEAM'] = game_results['WTEAM'].apply(lambda name: name.replace("St.", "Saint") if "St." in name else name)

game_results['LTEAM'] = game_results['LTEAM'].apply(lambda name: name.replace("St.", "State") if name == "Murray St." else name)
game_results['LTEAM'] = game_results['LTEAM'].apply(lambda name: name.replace("St.", "Saint") if "St." in name else name)
```

```python
# Make school names consistent between the two datasets
game_results['WTEAM'] = game_results['WTEAM'].apply(lambda name: "USC" if name == "Southern California" else name)
game_results['LTEAM'] = game_results['LTEAM'].apply(lambda name: "USC" if name == "Southern California" else name)

game_results['WTEAM'] = game_results['WTEAM'].apply(lambda name: "BYU" if name == "Brigham Young" else name)
game_results['LTEAM'] = game_results['LTEAM'].apply(lambda name: "BYU" if name == "Brigham Young" else name)

game_results['WTEAM'] = game_results['WTEAM'].apply(lambda name: "UCF" if name == "Central Florida" else name)
game_results['LTEAM'] = game_results['LTEAM'].apply(lambda name: "UCF" if name == "Central Florida" else name)

game_results['WTEAM'] = game_results['WTEAM'].apply(lambda name: "VCU" if name == "Virginia Commonwealth" else name)
game_results['LTEAM'] = game_results['LTEAM'].apply(lambda name: "VCU" if name == "Virginia Commonwealth" else name)

game_results['WTEAM'] = game_results['WTEAM'].apply(lambda name: "UAB" if name == "Alabama-Birmingham" else name)
game_results['LTEAM'] = game_results['LTEAM'].apply(lambda name: "UAB" if name == "Alabama-Birmingham" else name)

game_results['WTEAM'] = game_results['WTEAM'].apply(lambda name: "SMU" if name == "Southern Methodist" else name)
game_results['LTEAM'] = game_results['LTEAM'].apply(lambda name: "SMU" if name == "Southern Methodist" else name)

game_results['WTEAM'] = game_results['WTEAM'].apply(lambda name: "California" if name == "UC Berkeley" else name)
game_results['LTEAM'] = game_results['LTEAM'].apply(lambda name: "California" if name == "UC Berkeley" else name)

game_results['WTEAM'] = game_results['WTEAM'].apply(lambda name: "UTSA" if name == "Texas-San Antonio" else name)
game_results['LTEAM'] = game_results['LTEAM'].apply(lambda name: "UTSA" if name == "Texas-San Antonio" else name)

game_results['WTEAM'] = game_results['WTEAM'].apply(lambda name: "UMBC" if name == "Maryland-Baltimore County" else name)
game_results['LTEAM'] = game_results['LTEAM'].apply(lambda name: "UMBC" if name == "Maryland-Baltimore County" else name)

game_results['WTEAM'] = game_results['WTEAM'].apply(lambda name: name.replace(" (Cal.)", "") if name == "Saint Mary's (Cal.)" else name)
game_results['LTEAM'] = game_results['LTEAM'].apply(lambda name: name.replace(" (Cal.)", "") if name == "Saint Mary's (Cal.)" else name)

game_results['WTEAM'] = game_results['WTEAM'].apply(lambda name: name.replace(" (N.Y.)", "") if name == "Albany (N.Y.)" else name)
game_results['LTEAM'] = game_results['LTEAM'].apply(lambda name: name.replace(" (N.Y.)", "") if name == "Albany (N.Y.)" else name)

game_results['WTEAM'] = game_results['WTEAM'].apply(lambda name: name.replace("(Fla.)", "FL") if name == "Miami (Fla.)" else name)
game_results['LTEAM'] = game_results['LTEAM'].apply(lambda name: name.replace("(Fla.)", "FL") if name == "Miami (Fla.)" else name)

game_results['WTEAM'] = game_results['WTEAM'].apply(lambda name: name.replace("(Md.)", "MD") if name == "Loyola (Md.)" else name)
game_results['LTEAM'] = game_results['LTEAM'].apply(lambda name: name.replace("(Md.)", "MD") if name == "Loyola (Md.)" else name)

game_results['WTEAM'] = game_results['WTEAM'].apply(lambda name: name.replace("-", " ") if name == "Cal-Poly" else name)
game_results['LTEAM'] = game_results['LTEAM'].apply(lambda name: name.replace("-", " ") if name == "Cal-Poly" else name)

game_results['WTEAM'] = game_results['WTEAM'].apply(lambda name: name.replace("-", " ") if name == "Arkansas-Little Rock" else name)
game_results['LTEAM'] = game_results['LTEAM'].apply(lambda name: name.replace("-", " ") if name == "Arkansas-Little Rock" else name)

game_results['WTEAM'] = game_results['WTEAM'].apply(lambda name: name.replace("-", " ") if name == "Arkansas-Pine Bluff" else name)
game_results['LTEAM'] = game_results['LTEAM'].apply(lambda name: name.replace("-", " ") if name == "Arkansas-Pine Bluff" else name)

game_results['WTEAM'] = game_results['WTEAM'].apply(lambda name: name.replace("-", " ") if name == "Louisiana-Lafayette" else name)
game_results['LTEAM'] = game_results['LTEAM'].apply(lambda name: name.replace("-", " ") if name == "Louisiana-Lafayette" else name)

game_results['WTEAM'] = game_results['WTEAM'].apply(lambda name: "Loyola Chicago" if name == "Loyola (Ill.)" else name)
game_results['LTEAM'] = game_results['LTEAM'].apply(lambda name: "Loyola Chicago" if name == "Loyola (Ill.)" else name)

game_results['WTEAM'] = game_results['WTEAM'].apply(lambda name: "Southern Miss" if name == "Southern Mississippi" else name)
game_results['LTEAM'] = game_results['LTEAM'].apply(lambda name: "Southern Miss" if name == "Southern Mississippi" else name)

game_results['WTEAM'] = game_results['WTEAM'].apply(lambda name: "UT Arlington" if name == "Texas-Arlington" else name)
game_results['LTEAM'] = game_results['LTEAM'].apply(lambda name: "UT Arlington" if name == "Texas-Arlington" else name)

game_results['WTEAM'] = game_results['WTEAM'].apply(lambda name: name.replace("-", " ") if name == "Gardner-Webb" else name)
game_results['LTEAM'] = game_results['LTEAM'].apply(lambda name: name.replace("-", " ") if name == "Gardner-Webb" else name)

game_results['WTEAM'] = game_results['WTEAM'].apply(lambda name: f"{name} A&M" if name == 'Prairie View' else name)
game_results['LTEAM'] = game_results['LTEAM'].apply(lambda name: f"{name} A&M" if name == 'Prairie View' else name)

team_stats['TEAM'] = team_stats['TEAM'].apply(lambda name: "Long Island" if name == "LIU Brooklyn" else name)
                                              
team_stats['TEAM'] = team_stats['TEAM'].apply(lambda name: f"{name} State" if name == "Middle Tennessee" else name)
```

### Fix Errors in Dataset

```python
# Correct Auburn's seed from 4 to 5
game_results.iloc[68, 5] = 5

# Correct team name from Georgia State to Georgia Tech
game_results.iloc[37, 6] = 'Georgia Tech'
```

### Define New Features

```python
# Join Win % to game_results to use as determinant of better/worse seed when the seeds are equal
team_win_pct = team_stats[['YEAR', 'SEED', 'TEAM', 'WIN %']]
game_results = game_results.join(team_win_pct.set_index(['YEAR', 'SEED', 'TEAM']), on=['YEAR', 'WSEED', 'WTEAM'])
game_results = game_results.join(team_win_pct.set_index(['YEAR', 'SEED', 'TEAM']), on=['YEAR', 'LSEED', 'LTEAM'], lsuffix=' WTEAM', rsuffix=' LTEAM')
```

```python
# Define a team as the better-seeded team or the worse-seeded team & define the target: If an upset occurred
better_seeds = []
better_teams = []
worse_seeds = []
worse_teams = []
upsets = []

for row in game_results.iterrows():
    row = row[1]
    if row['WSEED'] == row['LSEED']:
        if row['WIN % WTEAM'] < row['WIN % LTEAM']:
            better_seeds.append(row['LSEED'])
            better_teams.append(row['LTEAM'])
            worse_seeds.append(row['WSEED'])
            worse_teams.append(row['WTEAM'])
            upsets.append(1)
        else:
            better_seeds.append(row['WSEED'])
            better_teams.append(row['WTEAM'])
            worse_seeds.append(row['LSEED'])
            worse_teams.append(row['LTEAM'])
            upsets.append(0)
    elif row['WSEED'] < row['LSEED']:
        better_seeds.append(row['WSEED'])
        better_teams.append(row['WTEAM'])
        worse_seeds.append(row['LSEED'])
        worse_teams.append(row['LTEAM'])
        upsets.append(0)
    else:
        better_seeds.append(row['LSEED'])
        better_teams.append(row['LTEAM'])
        worse_seeds.append(row['WSEED'])
        worse_teams.append(row['WTEAM'])
        upsets.append(1)

game_results['BETTER_SEED'] = better_seeds
game_results['BETTER_TEAM'] = better_teams
game_results['WORSE_SEED'] = worse_seeds
game_results['WORSE_TEAM'] = worse_teams
game_results['UPSET'] = upsets
```
